```python
import torch
adaptive_graphs = torch.load('classifier_crystal_graphs_v1.pt', weights_only=False)
fixed_graphs    = torch.load('classifier_crystal_graphs_v2.pt',   weights_only=False)

import pandas as pd

df_adaptive = pd.DataFrame([{
    'avg_degree': g.avg_degree if hasattr(g, 'avg_degree') else g.edge_index.shape[1] / g.x.shape[0],
    'n_edges': g.edge_index.shape[1],
    'isolated_nodes': (g.edge_index.shape[1] == 0),
    'cutoff_used': g.cutoff_used if hasattr(g, 'cutoff_used') else None
} for g in adaptive_graphs])

df_fixed = pd.DataFrame([{
    'avg_degree': g.edge_index.shape[1] / g.x.shape[0],
    'n_edges': g.edge_index.shape[1],
    'isolated_nodes': (g.edge_index.shape[1] == 0),
} for g in fixed_graphs])

print("=== Adaptive ===")
print(df_adaptive[['avg_degree', 'n_edges', 'isolated_nodes', 'cutoff_used']].describe().round(2))

print("\n=== Fixed 8Å ===")
print(df_fixed[['avg_degree', 'n_edges', 'isolated_nodes']].describe().round(2))

print(f"\nStructures with isolated nodes — Adaptive: {(df_adaptive['isolated_nodes'] > 0).sum()} "
      f"| Fixed 8Å: {(df_fixed['isolated_nodes'] > 0).sum()}")
print(f"Structures with avg_degree < 2  — Adaptive: {(df_adaptive['avg_degree'] < 2).sum()} "
      f"| Fixed 8Å: {(df_fixed['avg_degree'] < 2).sum()}")
```
